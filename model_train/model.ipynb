> Setup Environment dan Load & Cek Data


```python
import pandas as pd

path = "/content/drive/MyDrive/uas_ai_rafi/Testing/Heart_Disease_Prediction.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

> Memisahkan Fitur dan Labelnya

```python
X = df.drop("Heart Disease", axis=1)
y = df["Heart Disease"]
```

> Encoding Label (Presence / Absence)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

> Feature Scaling (Menormalkan skala data untuk KNN.)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

> Split Data Train & Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

> Training Model KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

> Evaluasi Model (Menilai Peforma dan Akurasi)

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = knn.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

> Cari Nilai K Terbaik

```python
for k in range(1, 8):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    acc = knn.score(X_test, y_test)
    print(f"K={k}, Accuracy={acc}")
```
